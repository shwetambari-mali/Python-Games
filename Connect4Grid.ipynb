```python
#Takes in player's marker decision
def player_input():
    marker = ''
    
    while marker not in ['X', 'O']:
        marker = input("Please pick a marker 'X' or 'O': ").upper()
        
    if marker == 'X':
        return('X', 'O')
    else:
        return('O', 'X')
```

```python
#Decides who plays first
import random
def random_choose_first():
    if random.randint(0, 1) == 1:
        return 'Player 1'
    else:
        return 'Player 2'
```

```python
#Takes in user position
def player_position_choice(board, player):
    position = 0
    
    while position not in range(1, 43) or not position_check(board, position):
        position = int(input(player + ' please enter a position between 1 to 42: '))
    return position
```

```python
#Checks if the position is free
def position_check(board, position):
    return board[position] == ' '
```

```python
#Marks the position
def position_marker(board, position, marker):
    board[position] = marker
```

```python
#Checks if board is full
def full_board_check(board):
    for i in range(1, 43):
        if position_check(board, i):
            return False
    return True
```

```python
#Asks user if they want to replay
def replay():
    return input('Do you want to replay? Yes or No: ').lower().startswith('y')
```

```python
#Game main
print('Welcomne to Connect4Grid')

while True:
    theBoard = [' '] * 43
    player1_marker, player2_marker = player_input()
    turn = random_choose_first()
    print(turn + ' will go first!')
    
    play_game = input('Are you ready to play? Enter Yes or No: ')
    if play_game.lower().startswith('y'):
        game_on = True
    else:
        game_on = False
        
    while game_on:
        if turn == 'Player 1':
            display_board(theBoard)
            position = player_position_choice(theBoard, turn)
            marked = position_marker(theBoard, position, player1_marker)
            
            if win_check(theBoard, player1_marker):
                display_board(theBoard)
                print('Congratulations! ' + turn + ' win!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 2'
        else:
            display_board(theBoard)
            position = player_position_choice(theBoard, turn)
            marked = position_marker(theBoard, position, player2_marker)
            
            if win_check(theBoard, player2_marker):
                display_board(theBoard)
                print('Congratulations! ' + turn + ' win!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print('The game is a darw!')
                    break
                else:
                    turn = 'Player 1'
                    
    if not replay():
        break
```
